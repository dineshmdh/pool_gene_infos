```python
# Creatd on Jan 3, 2018
# Process roadmaps metadata to map epigenome id to name and corresponding group. 
# The metadata can be downloaded here: http://egg2.wustl.edu/roadmap/web_portal/meta.html
import sys
reload(sys)
sys.setdefaultencoding('utf-8')  # helps with unicode issue
import pandas as pd

xl_file = "jul2013.roadmapData.qc.xlsx"
df_dict_meta = pd.read_excel(xl_file, sheetname=None, header=None)
print(df_dict_meta.keys()) # returns: [u'Consolidated_EpigenomeIDs_summa', u'Consolidated_EpigenomeIDs_QC', u'Unconsolidated_Release9_QC', u'BrowserInfo']
```

```python
df = df_dict_meta["Consolidated_EpigenomeIDs_summa"]
df.columns = df.iloc[0]  # set the first row as the header
df = df.drop(0, axis=0)  # drop the first row from the df
df = df.fillna("-")
df.head(3)
```

```python
# filter the columns
df = df[["Epigenome ID (EID)","Standardized Epigenome name", "Epigenome Mnemonic", "GROUP"]]
df = df[df["Epigenome ID (EID)"] != "-"]
df["EID_info"] = df["Epigenome ID (EID)"]+"_"+df["Epigenome Mnemonic"]+"_"+df["GROUP"]

print(df.shape)
df.head(3)
```

```python
df.to_csv("roadmap_EID_info.txt", sep="\t", header=True, index=False)
```
